```python
#create a calculator using python
import math as mh
class Calculator:
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2

    def add(self):
        return self.num1 + self.num2
    def subtract(self):
        return self.num1 - self.num2
    def multiply(self):
        return self.num1 * self.num2
    def divide(self):
        return self.num1 / self.num2
    def power(self):
        return self.num1 ** self.num2
    def square_root(self):
        return mh.sqrt(self.num1)
    def sine(self):
        return mh.sin(self.num1)


print("Welcome to the calculator")
print("Press space to continue for Two or more variable operation")
if(input()==" "):
    while(True):
      print("""
      Select an operation:
      1. Add
      2. Subtract
      3. Multiply
      4. Divide
      5.Power of that number in the form of x^y

    Enter 0 to exit and shut down calculator
""")
      num1 = int(input("Enter first number: "))
      num2 = int(input("Enter second number: "))
      choice_space = int(input("Enter your choice: "))
      calc = Calculator(num1, num2)
      if choice_space==1:
        print(f"Your Two number are {num1} and {num2} and the sum is {calc.add()}")
      elif choice_space==2:
        print(f"Your Two number are {num1} and {num2} and the difference is {calc.subtract()}")
      elif choice_space==3:
        print(f"Your Two number are {num1} and {num2} and the product is {calc.multiply()}")
      elif choice_space==4:
        print(f"Your Two number are {num1} and {num2} and the quotient is {calc.divide()}")
      elif choice_space==5:
        print(f"Your Two number are {num1} and {num2} and the power is {calc.power()}")
      elif choice_space==0:
        print("Shutting down calculator")
        break
      else:
        print("Invalid choice")
print("Okay.... so you want to do single operation press v to enter")
print("Press v to to single number operation ")
if(input()=="v"):
    print("""
    Select an operation:
    1. Square root
    2. Sine
    3. Cosine
    press 0 to exit and shut down calculator""")
    numv1=int(input("Enter the number"))
    while(True):
      calc=Calculator(numv1,0)
      choice_v=int(input("Enter your choice: "))
      if choice_v==1:
        print(f"The square root of {numv1} is {calc.square_root()}")
      elif choice_v==2:
        print(f"The sine of {numv1} is {calc.sine()}")
      elif choice_v==3:
        print(f"The cosine of {numv1} is {calc.cosine()}")
      elif choice_v==0:
        print("Shutting down calculator")
        break
      else:
        print("Invalid choice")


```
