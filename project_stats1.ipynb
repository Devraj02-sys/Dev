```python
# Import necessary libraries
import pandas as pd

# Load the datasets
math_df = pd.read_csv("/content/student-mat.csv", sep=';')
port_df = pd.read_csv("/content/student-por.csv", sep=';')

# Display the first few rows of each dataset
print("Mathematics Dataset:")
print(math_df.head())

print("\nPortuguese Language Dataset:")
print(port_df.head())

# Display basic information and data types
print("\nMathematics Dataset Info:")
math_df.info()

print("\nPortuguese Language Dataset Info:")
port_df.info()

# Display summary statistics for numeric columns
print("\nMathematics Dataset Statistics:")
print(math_df.describe())

print("\nPortuguese Language Dataset Statistics:")
print(port_df.describe())

# Explore unique values for categorical features
print("\nCategorical Value Exploration (Mathematics):")
for col in math_df.select_dtypes(include='object').columns:
    print(f"{col}: {math_df[col].unique()}")

print("\nCategorical Value Exploration (Portuguese):")
for col in port_df.select_dtypes(include='object').columns:
    print(f"{col}: {port_df[col].unique()}")
```

```python
# Import visualization libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Compute correlation matrix for numerical features (Mathematics)
math_corr = math_df.corr(numeric_only=True)

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(math_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix - Mathematics")
plt.show()

# Compute correlation matrix for numerical features (Portuguese)
port_corr = port_df.corr(numeric_only=True)

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(port_corr, annot=True, cmap='viridis', fmt=".2f")
plt.title("Correlation Matrix - Portuguese")
plt.show()
```
